# 1. Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.

```python
#Determine the distribution of each employees across each team
team_counts = df['Team'].value_counts()
print("The distribution of each employees across each team : \n", team_counts)
```

```python
#calculate the percentage split relative to the total number of employees

team_percentage = (team_counts/team_counts.sum())*100
#datframe for better visualization
distribution_df = pd.DataFrame({
    'Team' : team_counts.index,
    'number of employees' : team_counts.values,
    'percentage':team_percentage.values
   }   
)
print("The percentage split relative to the total number of employees : \n" , distribution_df)
```

**Graphical Representation**

```python
plt.figure(figsize=(20,8))
sns.barplot(x="Team",y="number of employees",data=distribution_df,hue="Team",palette = 'plasma')
plt.title("Number of employees by team")
plt.xticks(rotation=90)
plt.show()
```

**data story**

The barplot analysis shows that, the team 'New Orleans Pelicans' has highest number of employees that is 19.
and closely followed by 'Memphis Grizzlies' as 18.
Majority teams have 15 employees

# 4. Discover which team and position have the highest salary expenditure. 

```python
#missing values in each column
print(df.isnull().sum())
```

```python
#fill missing values in salary with mean
x= df['Salary'].mean()
df['Salary'].fillna(x,inplace=True)
df
```

```python
df['Salary'].value_counts()
```

```python
df_filterd = df[['Name','Team','Position','Salary']]
df_filterd
```

```python
#find maximum salary
total_salary = df.groupby(['Team','Position'])['Salary'].sum()
print(total_salary)
max_salary_data = total_salary.idxmax() # output ('Los Angeles Lakers', 'SF')
max_salary = total_salary.max()

print(f"Team and position have the highest salary is \n Team : {max_salary_data[0]} \n Position : {max_salary_data[1]} \n Highest salary : {max_salary}")
```

**Graphical Representation**

```python
total_salary = df.groupby(['Team','Position'])['Salary'].sum().reset_index()
#reset_index() is used to turn the resulting Series back into a DataFrame for easier plotting.
print(total_salary)
plt.figure(figsize=(20,8))
# Scatter plot
sns.scatterplot(data=total_salary, x='Team', y='Salary', hue='Position')

plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.title('Scatter plot position Vs Team with salary')
plt.xticks(rotation = 90)
plt.show()
```

**Data Story :**

 It reveals that the team Los Angeles Lakers SF position is the high salaried team 

# 5. Investigate if there's any correlation between age and salary, and represent it visually. 

**correlation**:  
measures the relationship between two variables. Values range from -1 to +1
Values close to +1 indicates strong positive correlation, -1 indicates a strong negative correlation and 0 indicates suggests no linear correlation.
Positive correlation variable move in same directions. As one increases, the other also increases.
Negative correlation variable move in opposite directions. An increase in one variable is associated with a decrease in the other.

**Heat Maps**
A heatmap is a graphical representation of data where individual values are represented by color gradients. It is a useful tool for visualizing the magnitude of values in a matrix or table format. Heatmaps are commonly used to display correlation matrices, data density, and other types of data where relationships and patterns can be better understood through color variations.

```python
df1 = df[['Age','Salary']]
df1
```

```python
corr = df1.corr()
corr
```

**Graphical representation**

```python
sns.heatmap(corr,cmap="YlGnBu",annot=True)
plt.title("Correlation matrix between age salary of employees")
plt.show()
```

**Data Story**
This correlation shows relationship between age and salary
Age vs Age (1.00000) :  This represents correlation of variable with itself, which is always 1 . It is positive correlation

Age vs Salary (0.21117) : This represents positive but relatively weak relationship between age and salary.This means that as age increases, there's a tendency for salaries to increase slightly, but the correlation isn't very strong.

Salary vs Age (0.21117): weak positive relationship between the two variables.
